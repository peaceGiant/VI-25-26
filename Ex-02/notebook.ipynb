```python
class FibNum:

    def __init__(self):
        self.fn2 = 1
        self.fn1 = 1

    def __next__(self): # next() is the heart of every iterator
                        # the next assignment is effective not only in terms
                        # of cutting lines from code, but also from the aspect of
                        # ensuring that we use only the old values of
                        # self.fn1 and self.fn2 in assignment
        (self.fn1, self.fn2, old_fn2) = (self.fn1 + self.fn2, self.fn1, self.fn2)
        return old_fn2

    def __iter__(self):
        return self
```

```python
f = FibNum()
for i in f:
    if i >= 1000:
        break
    print(i)

```

```python
class int_gen:

    def __init__(self):
        self.n = 1

    def __next__(self):
        old_n = self.n
        if old_n > 100:
            raise StopIteration

        self.n += 1
        return old_n

    def __iter__(self):
        return self
```

```python
g = int_gen()
for i in g:
    print(i)
```

```python
def g():
    n = 1
    while n <= 100:
        yield n
        n += 1
```

```python
iter_ = g()
```
